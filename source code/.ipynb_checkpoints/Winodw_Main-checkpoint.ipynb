```python
import cv2
import sys
from PyQt5 import *
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from drag_gray import *
from draw_figure import *
from blur import *
from flip import *
from bright_track import *
from Window_Sub import SubWindow

class MainWindow(QMainWindow):
    def __init__(self):                                                                 #self : 클래스 내 호출한 인스턴스 & __init__ 클래스를 통해 인스턴스 생성 시 항상 실행되는 부분
        super().__init__()
        self.setGeometry(500,200,700,700)
        self.setWindowTitle('Photoshop')
        self.setWindowIcon(QIcon('img\icon\smile.png'))
        self.initUI()
    
    def initUI(self):
        self.date = QDate.currentDate()
        self.statusBar().showMessage(self.date.toString(Qt.DefaultLocaleLongDate))      # 윈도우 창에 현재 날짜 출력
        menubar = self.menuBar()                                                        # 메뉴바 생성
        menu = menubar.addMenu("HOME")
        menu_FILE = menubar.addMenu("FILE")
        menu_EDIT = menubar.addMenu("EDIT")
        menu_COLOR = menubar.addMenu("COLOR")
        menu_DRAWING = menubar.addMenu("DRAWING")
        # FILE
        New_action=QAction(QIcon('img\icon\plus.png'),'NEW',self)                       # QAction()으로 클릭 시 action으로 연결되는 것 표시
        New_action.triggered.connect(self.new_file)                                     # QAction()으로 클릭 시 연결되는 action를 함수와 연결
        menu_FILE.addAction(New_action)                                                 # 연결한 액션을 FILE에 action추가
        Exit_action=QAction(QIcon('img\icon\exit.png'),'EXIT',self)
        Exit_action.triggered.connect(QCoreApplication.instance().quit)
        menu_FILE.addAction(Exit_action)
        # EDIT
        Flip_action=QAction(QIcon('img\icon\sflip.png'),'FLIP',self)
        Flip_action.triggered.connect(self.flip)
        menu_EDIT.addAction(Flip_action)
        Blur_action=QAction(QIcon('img\icon\pblur.png'),'BLUR',self)
        Blur_action.triggered.connect(self.bluring)
        menu_EDIT.addAction(Blur_action)
        Bright_action=QAction(QIcon('img\icon\sun.png'),'BRIGHT',self)
        Bright_action.triggered.connect(self.bright)
        menu_EDIT.addAction(Bright_action)
        # COLOR
        gray_action=QAction(QIcon('img\icon\pen.png'),'DRAG_GRAY',self)
        gray_action.triggered.connect(self.drag_gray)
        menu_COLOR.addAction(gray_action)
        color_action=QAction(QIcon('img\icon\pen2.png'),'COLOR',self)
        color_action.triggered.connect(self.color)
        menu_COLOR.addAction(color_action)
        # DRAW
        draw_action=QAction(QIcon('img\icon\erec.png'),'DRAW',self)
        draw_action.triggered.connect(self.draw_figure)
        menu_DRAWING.addAction(draw_action)
        
    def new_file(self):                             # 폴더에서 파일 선택해서 불러오는 함수
        fname = QFileDialog.getOpenFileName(self)   # QFileDialog : 파일 또는 경로를 선택할 수 있도록 하는 다이얼로그
        if fname[0]:                                # fname[0], 즉 파일이 선택되었을 경우
            self.label = QLabel(self)               # QLabel를 통해 윈도우 창에 보이도록 표시
            pixmap=QPixmap(fname[0])                # QPixmap()를 통해 이미지 불러오기
            self.label.setPixmap(pixmap)
            self.label.setGeometry(100,30,500,620)  # 출력되는 이미지 크기 지정
            self.label.show()                       # 이미지 출력
            
    def drag_gray(self):  
        gray()
    def draw_figure(self):
        draw()
    def flip(self):
        flip()
    def bluring(self):
        blur()
    def color(self):                    #Window_Sub와 연결
        self.w = SubWindow()
        self.w.show()
    def bright(self):
        trackbar()    
def main():
    app = QApplication(sys.argv)        #응용 프로그램을 작동시키기 위한 QApplication()
    mainWindow = MainWindow()
    mainWindow.show()
    sys.exit(app.exec())                # app  객체를 실행시키고 종료시킨다
        
if __name__ == '__main__':              # 인터프리터에서 직접 실행했을 경우에만 if문 내 코드 실행
    main()
```
